```python
%%capture
import struct 
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

from helpers import read_echogram
```

```python
class EchoGram:
    def __init__(self, filename, channels=None, depthline=False):
        self.filename = filename
        self.data, self.echo = self.test_overlay_wb_echo(filename, channels, depthline)
        
    def test_overlay_wb_echo(self, filename, channels=None, depthline=False):
        # Load sounding data and echogram.
        data, echo = read_echogram(filename, channels)

        ## Calculate sampling rate, using assumption that y-axis is in depth (ft)
        ## - not actually used in this function but given here as an illustration.
        # time1 = np.array([i["time1"] for i in data])
        # dz = time1[1:] - time1[:-1]
        if depthline:
            ## Calculate the water depth in y-axis samples
            wb = np.array([i["water_depth"] for i in data])
            lower_limit = np.array([i["lower_limit"] for i in data])
            wb_s =  wb * echo.shape[0] / lower_limit

        fig, axes = plt.subplots()
        axes.imshow(echo, aspect='auto')
        if depthline:
            axes.plot([i["frame_index"] for i in data],
                        wb_s, linewidth=2, color="red")
        fig.show()
        #fig.savefig('sss.png', dpi=fig.dpi)
        return data, echo
```

### Channel (sonar sensor type)
- 0 = Primary (Traditional Sonar)
- 1 = Secondary (Traditional Sonar)
- 2 = DSI (Downscan Imaging)
- 3 = Sidescan Left
- 4 = Sidescan Right
- 5 = Sidescan (Composite)
Another other value is treated as Invalid.

```python
def dir_comparison(no_so, ea_we):
    avg_lon_no_so = np.mean(df2.loc[no_so,:]['longitude'])
    avg_lat_ea_we = np.mean(df2.loc[ea_we,:]['latitude'])
    ea_we_in = np.argmin(np.abs(df2.loc[ea_we,:]['longitude']-avg_lon_no_so))
    no_so_in = np.argmin(np.abs(df2.loc[no_so,:]['latitude']-avg_lat_ea_we))

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    no_so_depth = df2.loc[no_so,:]['water_depth_m']
    ea_we_depth = df2.loc[ea_we,:]['water_depth_m']
    ax.scatter(range(len(no_so_depth)), no_so_depth, c='blue', label='North-South')
    ax.scatter(range(len(ea_we_depth)), ea_we_depth, c='red', label='East-West')
    min_depth = np.min([np.min(no_so_depth), np.min(ea_we_depth)])
    max_depth = np.max([np.max(no_so_depth), np.max(ea_we_depth)])
    ax.plot([ea_we_in]*2,[min_depth,max_depth], 'r', label='East-West Intersection')
    ax.plot([no_so_in]*2,[min_depth,max_depth], 'b', label='North-South Intersection')
    ax.set_ylabel('depth [m]')
    ax.set_xlabel('Data points')
    ax.legend()
    plt.show()
```

```python
no_so1 = range(6140,6200)
ea_we1 = range(15700,15750)
dir_comparison(no_so1, ea_we1)
```

```python
no_so2 = range(4790,4860)
ea_we2 = range(15600,15650)
dir_comparison(no_so2, ea_we2)
```
